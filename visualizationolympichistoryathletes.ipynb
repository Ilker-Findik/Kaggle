```python
df = pd.read_csv("/kaggle/input/120-years-of-olympic-history-athletes-and-results/athlete_events.csv")
```

```python
df.info()
```

```python
plt.scatter("Height","Weight",data = df)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athlete Height - Weight")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x = "Height",y = "Weight",data = df)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Height - Weight")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x = "Height", y = "Weight", hue = "Sex", data = df)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Height - Weight")
plt.show()
```

```python
df["Medal"].unique()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x = "Height", y = "Weight", hue = "Sex", style = "Medal", data = df)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Atheletes Height - Weight")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x = "Height", y = "Weight", hue = "Sex",data = df)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Atheletes Height - Weight")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.displot(x = "Height", hue = "Sex",data = df)
plt.ylabel("Frequency")
plt.title("Atheletes Height Histogram")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.displot(x = "Height", hue = "Sex",data = df, kind="kde")
plt.ylabel("Frequency")
plt.title("Atheletes Height Histogram")
plt.show()
```

```python
sns.barplot(x = "Medal", y = "Height", hue = "Sex", data = df)
plt.title("Medals by Height")
plt.show()
```

```python
sns.catplot(x = "Medal", y = "Height", hue = "Sex", col="Season", data = df)
plt.title("Medals by Height")
plt.show()
```
